### CADD

* “General” prediction score, incorporates DNA/protein sequence features as well as epigenomic signals and provide deleteriousness predictions for any SNV in the human genome, coding or non-coding.

### fathmm-MKL_score

* “General” prediction score, incorporates DNA/protein sequence features as well as epigenomic signals and provide deleteriousness predictions for any SNV in the human genome, coding or non-coding.

### fitCons: integrated_fitCons_score,GM12878_fitCons_score,H1-hESC_fitCons_score, HUVEC_fitCons_score

* “General” prediction score, incorporates DNA/protein sequence features as well as epigenomic signals and provide deleteriousness predictions for any SNV in the human genome, coding or non-coding.
* Predictions at the genome position level NOT snv

### DANN_score

* “General” prediction score, incorporates DNA/protein sequence features as well as epigenomic signals and provide deleteriousness predictions for any SNV in the human genome, coding or non-coding.

### MetaSVM_score

* ENSEMBLE score based on 10 component scores (SIFT, PolyPhen-2 HDIV, PolyPhen-2 HVAR, GERP++, MutationTaster, Mutation Assessor, FATHMM, LRT, SiPhy, PhyloP) and the maximum frequency observed in the 1000 genomes populations, second highest performer in separation power.

### MetaLR_score

* ENSEMBLE score based on 10 component scores (SIFT, PolyPhen-2 HDIV, PolyPhen-2 HVAR, GERP++, MutationTaster, Mutation Assessor, FATHMM, LRT, SiPhy, PhyloP) and the maximum frequency observed in the 1000 genomes populations, highest performer in separation power.
